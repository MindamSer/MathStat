```python
import numpy as np
import matplotlib.pyplot as plt
import random

Teta = 300
N = 100

x = np.array([random.randint(1, Teta) for i in range(N)])

def teta1(v):
    return 2 * np.mean(v)

def teta2(v):
    return np.min(v) * (len(v) + 1)

def teta3(v):
    return np.max(v) * (len(v) + 1) / len(v)

def teta4(v):
    return np.min(v) + np.max(v)

def teta5(v):
    return v[0] + np.mean(v[1::])
```

```python
tetas = [teta1, teta2, teta3, teta4, teta5]
dts = [dt1, dt2, dt3, dt4, dt5]

for i in range(len(dts)):
    dts[i] = np.array([0. for i in range(N)])

for i in range(N):
    for j in range(len(tetas)):
        dts[j][i] = np.abs(tetas[j](x[0:i+1:]) - Teta)/Teta

fig, ax = plt.subplots(5,1,figsize=(9, 45))
fig.suptitle("|\u03B8\u0303'-\u03B8|", fontsize=20)
label = ["\u03B81","\u03B82","\u03B83","\u03B84","\u03B85"]

for i in range(len(dts)):
    ax[i].set_title(label[i])
    ax[i].minorticks_on()
    ax[i].grid(which='major', color = 'k', linewidth = 1)
    ax[i].grid(which='minor', color = 'k', linestyle = ':')
    ax[i].set_xlabel('n')
    ax[i].set_ylabel('\u0394\u03B8')
    ax[i].scatter(np.linspace(1,N,N),dts[i])
    ax[i].plot(np.linspace(1,N,N),dts[i])
```
